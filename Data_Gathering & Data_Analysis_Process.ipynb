# to_csv

```python
df=pd.read_csv("D:\\Python_Notes\\deliveries.csv")
df
```

```python
df1=df.groupby('batsman')['batsman_runs'].sum().reset_index()
df1
```

```python
df1.to_csv('batsman_runs.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

# to_excel

```python
df2=df.groupby('batsman')['batsman_runs'].sum().reset_index()
df2
```

```python
df2.to_excel('batsman_runs.xlsx')
```

```python
df2.to_excel('batsman_runs.xlsx',sheet_name='batsman_runs')
```

```python
df3=df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

# Make multiple sheets in excel

```python
with pd.ExcelWriter('batsman_runs.xlsx') as writer:
    df2.to_excel(writer,sheet_name='batsman_runs')
    df3.to_excel(writer,sheet_name='batsman_runs1')
    
```
